```python
import numpy as np
from skimage import io
from skimage import exposure
from skimage import color

import supervisely_lib as sly

import os

from skimage.transform import rescale

from matplotlib import pyplot as plt
```

```python
images_dir = './data'
```

```python
img_paths = sly.fs.list_files(images_dir)
```

```python
img_paths
```

```python
image = io.imread(img_paths[0])

plt.imshow(image[50,...], cmap = 'gray')
plt.show()
```

```python
plt.hist(image.ravel(), bins = 256)
plt.show()
```

```python
resc = exposure.rescale_intensity(image, in_range='image', out_range=np.uint8)
```

```python
plt.hist(resc.ravel(), bins = 256)
plt.show()
```
